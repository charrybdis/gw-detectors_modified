```python
from gwdetectors import *
import matplotlib.pyplot as plt
```

```python
detector = TwoArmDetector('test', PSDS[KNOWN_PSDS[1]], (0, 0, 0), arms = ((2e-5, 0, 0), (0, 2e-5, 0)))
#detector = DETECTORS['L_aplus-design']

fsr = 1 / (2 * np.linalg.norm(detector.arms[0]))
```

```python
res = 180
fdivfsr_max = 2.0 
polarization = 3*np.pi/4

phi = np.linspace(-np.pi, np.pi, res)
theta = np.linspace(0, np.pi,res)
Phi, Theta = np.meshgrid(phi, theta)

phi_flat = Phi.flatten()
theta_flat = Theta.flatten()

psi = np.full(np.shape(phi_flat), polarization)

#frq_max= fdivfsr_max * fsr 
frq_max=0.8
freqs = np.linspace(0, frq_max, 10)

F_plus, F_cross, F_x, F_y, F_b, F_l = detector._geographic_unphased_response(freqs, phi_flat, theta_flat, psi, 
                                                                             detector.arm_directions, 
                                                                             long_wavelength_approximation=False)
```

```python
fr = 0

Fp = F_plus[fr].reshape((res, res))
Fc = F_cross[fr].reshape((res, res))
Fx = F_x[fr].reshape((res, res))
Fy = F_y[fr].reshape((res, res))
Fb = F_b[fr].reshape((res, res))
Fl = F_l[fr].reshape((res, res))
```

```python
# Create coordinate grid for plot (phi in [-pi, pi], theta in [-pi/2, pi/2])
phi = np.linspace(-np.pi, np.pi, res)
theta = np.linspace(-np.pi/2., np.pi/2.,res)
Phi, Theta = np.meshgrid(phi, theta)
```

### Plots of magnitude

```python
data = np.sqrt(np.abs(Fc) ** 2 + np.abs(Fp) ** 2)

# Plot figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='mollweide')
im = ax.pcolormesh(Phi,Theta, data, cmap=plt.cm.Reds) 
plt.colorbar(im)
plt.show()
#plt.savefig(f'Fc_sq+Fp_sq_{fr}.png')
```
