#### Project

```python
# Pick shape of sine-Gaussian signal pulse
a = fsr # frequency of sinusoidal component 
A = 1e-23 # Gaussian peak amplitude
c = 1e-4 # standard deviation of Gaussian envelope
dt= 0

numpts=120
spread=4

freqs_signal, ast_signal = Functions.produce_freqs_signal(numpts, spread, a, A, c, dt)
```

```python
azim = np.array([0.261799, 0.261799, 0.261799, 0.261799])
pole = np.array([0, np.pi/6, np.pi/3, np.pi/2])
psi = 0
geocent = 0

project_p = [detector.project(freqs_signal, geocent, azim[i], pole[i], psi, coord='geographic', hp=ast_signal)
             for i in range(len(azim))]
project_x = [detector.project(freqs_signal, geocent, azim[i], pole[i], psi, coord='geographic', hx=ast_signal)
             for i in range(len(azim))]
project_vx = [detector.project(freqs_signal, geocent, azim[i], pole[i], psi, coord='geographic', hvx=ast_signal)
             for i in range(len(azim))]
project_vy = [detector.project(freqs_signal, geocent, azim[i], pole[i], psi, coord='geographic', hvy=ast_signal)
             for i in range(len(azim))]
project_b = [detector.project(freqs_signal, geocent, azim[i], pole[i], psi, coord='geographic', hb=ast_signal)
             for i in range(len(azim))]
project_l = [detector.project(freqs_signal, geocent, azim[i], pole[i], psi, coord='geographic', hl=ast_signal)
             for i in range(len(azim))]
```

```python
colours = ['b', 'g', 'r', 'c']

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.set_ylabel('Magnitude')
for i in range(4): 
    ax1.plot(freqs_signal/fsr, np.abs(project_p[i]), colours[i])
ax2.set_ylabel('Phase (rad)')
ax2.set_xlabel('f/f_fsr')
for i in range(4): 
    ax2.plot(freqs_signal/fsr, np.arctan2(project_p[i].imag, project_p[i].real), colours[i])
plt.show()
```

```python
colours = ['b', 'g', 'r', 'c']

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.set_ylabel('Magnitude')
for i in range(4): 
    ax1.plot(freqs_signal/fsr, np.abs(project_x[i]), colours[i])
ax2.set_ylabel('Phase (rad)')
ax2.set_xlabel('f/f_fsr')
for i in range(4): 
    ax2.plot(freqs_signal/fsr, np.arctan2(project_x[i].imag, project_x[i].real), colours[i])
plt.show()
```

```python
colours = ['b', 'g', 'r', 'c']

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.set_ylabel('Magnitude')
for i in range(4): 
    ax1.plot(freqs_signal/fsr, np.abs(project_vx[i]), colours[i])
ax2.set_ylabel('Phase (rad)')
ax2.set_xlabel('f/f_fsr')
for i in range(4): 
    ax2.plot(freqs_signal/fsr, np.arctan2(project_vx[i].imag, project_vx[i].real), colours[i])
plt.show()
```
