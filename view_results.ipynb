```python
names = ['1_1e-3', '1_1e-2', '2_1e-2', '0.5_1e-3', '1_5e-3', '1_8e-3', '0.4_1e-3', '0.8_1e-3', '1.2_1e-3', 
         '1.6_1e-3', '2.0_1e-3', '1_8e-3', '0.4_5e-3', '0.8_5e-3', '1.2_5e-3', '1.6_5e-3', 
         '2.0_5e-3', '0.4_1e-2', '0.8_1e-2', '1.2_1e-2', '1.6_1e-2', '2.0_1e-2']
freq_paths = []
info_paths = []
for name in names:
    freq_paths.append('/fs/lustre/cita/jewelcao/plots/hp_true/hb/' + name)
    info_paths.append('/fs/lustre/cita/jewelcao/results/hp_true/hb/' + name + '/info.pickle')
```

```python
CE_match_avgs = []
CE_match_meds = []
HLVK_match_avgs = []
HLVK_match_meds = []

for path in freq_paths:
    with open(f"{path}/summary_stats.pickle", "rb") as f: 
        stats = pickle.load(f)
        CE_match_avg = stats['CE']['average CE match']
        CE_match_med = stats['CE']['median CE match']
        HLVK_match_avg = stats['excluding CE']['average match excluding CE']
        HLVK_match_med = stats['excluding CE']['median match excluding CE']
        CE_match_avgs.append(CE_match_avg)
        CE_match_meds.append(CE_match_med)
        HLVK_match_avgs.append(HLVK_match_avg)
        HLVK_match_meds.append(HLVK_match_med)
freqs = []
stddevs = []
for path in info_paths:
    with open(path, "rb") as f:
        info = pickle.load(f)
        freq = info['a']
        stddev = 1 / (2 * np.pi * info ['c'])
        print(info['c'])
        freqs.append(freq)
        stddevs.append(stddev)
```

```python
# Average of HLVK matches for breathing mode, plotted on frequency and standard deviation of the signal

cm = plt.cm.get_cmap('RdYlBu')
sc = plt.scatter(freqs, stddevs, c=HLVK_match_avgs, cmap=cm)
plt.colorbar(sc)
plt.xlabel('Frequency Center (Hz)')
plt.ylabel('Standard Deviation (Hz)')
plt.show()
```

```python
# Average of CE matches for breathing mode, plotted on frequency and standard deviation of the signal

cm = plt.cm.get_cmap('RdYlBu')
sc = plt.scatter(freqs, stddevs, c=CE_match_avgs, cmap=cm)
plt.colorbar(sc)
plt.xlabel('Frequency Center (Hz)')
plt.ylabel('Standard Deviation (Hz)')
plt.show()
```
